```python
wells_info_df['CompletionDate'] = pd.to_datetime(wells_info_df['CompletionDate'])
wells_info_df['FirstProductionDate'] = pd.to_datetime(wells_info_df['FirstProductionDate'])
diff = np.floor((wells_info_df['CompletionDate'] - wells_info_df['FirstProductionDate']).dt.days / 30)
diff[diff < 0] = 0
diff_df = pd.DataFrame(diff,columns = ['Prod_duration'])
pd.concat([wells_info_df['API'],diff], axis=1)
```

3.Fill

```python
wells_info_na_df = pd.read_csv('wells_info_na.csv')
wells_info_na_df 
```

```python
wells_info_na_df = wells_info_na_df.fillna(wells_info_na_df.median())
columns = list(wells_info_na_df)
for i in columns:
    wells_info_na_df[i] = wells_info_na_df[i].fillna(wells_info_na_df[i].value_counts().index[0])
wells_info_na_df
```

```python
production_df = pd.read_csv('production.csv')
```

```python
tmp = production_df.groupby('API').sum()[['Liquid','Gas','Water']].sum(axis = 1)
df_summ = pd.DataFrame(tmp,columns = ['Debit'])
groups = pd.DataFrame(production_df.groupby('API').head(12))
tmp = groups.groupby('API').sum()[['Liquid','Gas','Water']].sum(axis = 1)
df_summ_12m = pd.DataFrame(tmp,columns = ['Debit_12m'])
wells_info_df = pd.merge(wells_info_df,df_summ,on='API')
pd.merge(wells_info_df,df_summ_12m,on='API')
```
